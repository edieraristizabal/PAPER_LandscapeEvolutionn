# Distance to Border

```python
import numpy as np
import pandas as pd
import geopandas as gpd
import seaborn as sbn
import matplotlib
import matplotlib.pyplot as plt
import rasterio as rio
```

### Recent landslides

```python
rec=gpd.read_file("../Data/recent2_chidistance.gpkg")
rec.plot()
rec.head(2)
```

```python
f, ax = plt.subplots(figsize=(10,5))
ax.hist(rec["distance"]/1000,bins=20,color="k",alpha=0.8,width=1.05,label="Recent landslides",weights=np.full(len(rec["distance"]), 1/len(rec["distance"])))
ax.set_xlabel('Distance (km)', fontsize=20)
ax.set_ylabel('Frequency', fontsize=20)
#ax.set(xticks=ax.get_xticks()[::2], xticklabels=ax.get_xticklabels()[::2]);
ax.set(yticks=ax.get_yticks()[::2], yticklabels=ax.get_yticklabels()[::2]);
ax.xaxis.set_tick_params(labelsize=18)
ax.yaxis.set_tick_params(labelsize=18)
#ax.set_xlim(0,35)
#ax.set_ylim(0,2200)
ax.spines[['right', 'top','left','bottom']].set_linewidth(1)
f.tight_layout() 
plt.legend(frameon=False,loc='upper right', prop={'size': 20},handlelength=0)
```

```python
raster = rio.open("G:/My Drive/INVESTIGACION/POSDOC/Data/Raster/distance_NOrecent2.tif")
ran_f = raster.read(1)
```

```python
ran_f=np.where(ran_f<0,np.nan,ran_f)
plt.imshow(ran_f)
```

```python
ran=np.array(ran_f).flatten()
ran = ran[~np.isnan(ran)]
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.hist(ran,orientation='vertical',color='red',alpha=0.6,bins=20,label="Recent frequency",weights=np.full(len(ran), 1/len(ran)))
ax.set_ylabel('Elevation (m)', fontsize=20,fontweight='bold')
```

```python
#se demora 10 min
serie=rec["distance"]
a=b=c=d=e=f=g=h=i=j=0
for row in serie:
        if (0 <= row) & (row < 5000):
          a+=1
        elif (5000 <= row) & (row < 10000):
          b+=1
        elif (10000 <= row) & (row < 15000):
          c+=1
        elif (15000 <= row) & (row < 20000):
          d+=1
        elif (20000 <= row) & (row < 25000):
          e+=1
        elif (25000 <= row) & (row < 30000):
          f+=1
        elif (30000 <= row) & (row < 35000):
          g+=1
        elif (35000 <= row) & (row < 40000):
          h+=1
        elif (40000 <= row) & (row < 45000):
          i+=1
        if (row>=45000):
          j+=1

total=[a,b,c,d,e,f,g,h,i,j]
print(total)

#ran_serie=pd.Series[ran]
a=b=c=d=e=f=g=h=i=j=0
for row in ran:
        if (0 <= row) & (row < 5000):
          a+=1
        elif (5000 <= row) & (row < 10000):
          b+=1
        elif (10000 <= row) & (row < 15000):
          c+=1
        elif (15000 <= row) & (row < 20000):
          d+=1
        elif (20000 <= row) & (row < 25000):
          e+=1
        elif (25000 <= row) & (row < 30000):
          f+=1
        elif (30000 <= row) & (row < 35000):
          g+=1
        elif (35000 <= row) & (row < 40000):
          h+=1
        elif (40000 <= row) & (row < 45000):
          i+=1
        if (row>=45000):
          j+=1

total1=[a,b,c,d,e,f,g,h,i,j]
print(total1)
```

```python
fig, ax = plt.subplots(figsize = (7,5))
rece=100*(total/np.array(total).sum())
rando=100*(total1/np.array(total1).sum())
range=[0,5,10,15,20,25,30,35,40,45]
ax.bar(range,rece/rando,width=4,color="k",alpha=0.6,label="Recent landslides")
ax.set_ylabel('Relative frequency',fontsize=20)
ax.set_xlabel('Distance to transient divides (km)', fontsize=20)
ax.xaxis.set_tick_params(labelsize=18)
ax.yaxis.set_tick_params(labelsize=18)
ax.hlines(y=1.0, xmin=-1.0, xmax=40.0, color='black',linestyles='dotted')
ax.set_xlim(-2,40)
ax.set_ylim(0,2.5)
ax.spines[['right','top','left','bottom']].set_linewidth(1)
fig.tight_layout() 
ax.text(0,2.3,"(D)",fontsize=16,weight='bold')
plt.legend(frameon=False,loc='upper right',prop={'size': 20},handlelength=0)
plt.savefig("../Figures/fig10D_chi.png", dpi=500)
```

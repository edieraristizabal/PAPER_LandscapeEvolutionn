### Code for Figure 2 A, B, and C

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import geopandas as gpd
```

```python
p  = np.genfromtxt("G:/My Drive/INVESTIGACION/POSDOC/Perfiles/NS/statslines_NS32.txt")
fig, ax = plt.subplots(figsize=(25,10));
ax.plot(p[:,0]/1000,p[:,1],'-',c='gray',lw=1);
ax.plot(p[:,0]/1000,p[:,3],'-k',lw=1.5);
plt.plot(p[:,0]/1000,p[:,2],'-',c='gray',lw=1);
x=p[:,0]/1000
y1=p[:,1]
y2=p[:,2]
ax.fill_between(x,y1,y2, facecolor='r', alpha=0.4)
ax.set_xlim(0,250)
ax.set_ylim(0,4000)
ax.set_yticks([0,1000,2000,3000,4000])
plt.ylabel('Elevation (m)',size=40);
ax.tick_params(labelsize=30)
#plt.xlabel('Distance along profile (km)',size=40);
plt.xticks(size=30);

r  = pd.read_csv("G:/My Drive/INVESTIGACION/POSDOC/Hydrometereological/dateA.csv")
ax2=ax.twinx()
ax2.set_ylim(1000,8500)
ax2.set_ylabel('Rainfall (mm/year)', fontsize=36)
ax2.tick_params(labelsize=30)
ax2.set_yticks([2000,4000,6000,8000])
ax2.invert_yaxis()
ax2.bar(np.linspace(min(p[:,0])/1000,max(p[:,0])/1000,5000),r["first"], color='tab:blue',alpha=0.05)
ax.text(10,3500,'(A)',fontsize = 34,fontweight='bold',color ="black") 
fig.savefig("G:/My Drive/INVESTIGACION/POSDOC/Figuras/AOI_rainfall_A.png", dpi=500);
```
